# 3. Sensitivity of r and alpha

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# 修正后的数据准备
data = {
    (2,1): {'train': 0.3292, 'val': 0.3187, 'acc': 0.8969},
    (2,2): {'train': 0.3072, 'val': 0.3081, 'acc': 0.8984},
    (2,3): {'train': 0.3041, 'val': 0.3084, 'acc': 0.8938},
    (6,1): {'train': 0.3013, 'val': 0.3073, 'acc': 0.9000},
    (6,2): {'train': 0.2989, 'val': 0.3071, 'acc': 0.9016},
    (6,3): {'train': 0.3119, 'val': 0.3175, 'acc': 0.8922},
    (10,1): {'train': 0.3114, 'val': 0.3171, 'acc': 0.8922},
    (10,2): {'train': 0.3092, 'val': 0.3172, 'acc': 0.8953},
    (10,3): {'train': 0.3079, 'val': 0.3169, 'acc': 0.8984}
}

# 参数范围
r_values = sorted({k[0] for k in data.keys()})  # [2, 6, 10]
alpha_values = sorted({k[1] for k in data.keys()})  # [2, 4, 6, 10, 12, 20, 30]

# 创建矩阵的通用函数
def create_matrix(metric):
    matrix = np.zeros((len(r_values), len(alpha_values)))
    for i, r in enumerate(r_values):
        for j, alpha in enumerate(alpha_values):
            matrix[i,j] = data.get((r, alpha), {}).get(metric, np.nan)
    return matrix

# 创建颜色映射
loss_cmap = LinearSegmentedColormap.from_list('loss', ['#2ecc71', '#f1c40f', '#e74c3c'])
acc_cmap = LinearSegmentedColormap.from_list('acc', ['#e74c3c', '#f1c40f', '#2ecc71'])

# 增强的热力图绘制函数
def plot_enhanced_heatmap(matrix, title, cmap, ax=None):
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 6))
    
    # 计算显示范围
    vmin = np.nanmin(matrix)
    vmax = np.nanmax(matrix)
    
    # 设置mask隐藏NaN值
    mask = np.isnan(matrix)
    
    sns.heatmap(
        matrix, 
        annot=True, 
        fmt=".4f",
        cmap=cmap,
        mask=mask,
        ax=ax,
        cbar_kws={'label': title},
        vmin=vmin-(vmax-vmin)*0.1,  # 扩展10%范围
        vmax=vmax+(vmax-vmin)*0.1,
        xticklabels=alpha_values,
        yticklabels=r_values,
    )
    
    ax.set_xlabel('LoRA Alpha Multiplier', fontsize=20)
    ax.set_ylabel('Rank (r)', fontsize=20)
    ax.set_title(f'{title}\n(dropout=0.1, bias=all)', pad=15, fontsize=20)
    
    # 添加网格线
    for _, spine in ax.spines.items():
        spine.set_visible(True)
        spine.set_edgecolor('gray')
    
    return ax

# 生成矩阵
train_loss = create_matrix('train')
print(train_loss)
val_loss = create_matrix('val')
val_acc = create_matrix('acc')

# 创建三合一可视化
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# 绘制三个热力图
plot_enhanced_heatmap(train_loss, 'Training Loss', loss_cmap, axes[0])
plot_enhanced_heatmap(val_loss, 'Validation Loss', loss_cmap, axes[1])
plot_enhanced_heatmap(val_acc, 'Validation Accuracy', acc_cmap, axes[2])

plt.tight_layout(pad=3)
plt.savefig('combined_heatmaps.png', dpi=300, bbox_inches='tight')
plt.show()
```
